Zadanie polega na policzeniu średnich miesięcznych w sposób "poprawny", tj. średnią trzeba ważyć za pomocą niepewności pomiarowych. Proszę zastosować funkcję gropupby. 


```python
import pandas as pd
```

```python
tabela = pd.read_csv('data-errors.csv', skiprows=2)
tabela = tabela.drop('Unnamed: 0',1)
```

```python
d = tabela['Rok']
tabela.Rok = d.fillna(method='ffill')

from datetime import date, timedelta
def week_start_date(year, week, dow=0):    
    """
    :param int year: Numer roku
    :param int week: Numer tygodnia w roku 
    :param int dow: Funkcja ta zwraca datę, zwróci datę dnia o indeksie dow, zero oznacza poniedziałek, jeden wtorek itp. 
    """
    # http://stackoverflow.com/a/1287862/7918
    d = date(year, 1, 1)    
    delta_days = d.isoweekday() - 1
    delta_weeks = week
    if year == d.isocalendar()[0]:
        delta_weeks -= 1
    delta = timedelta(days=-delta_days+dow, weeks=delta_weeks)
    return d + delta
```

```python
tabela["Date"] = [week_start_date(int(x), int(y)) for x,y in zip(tabela['Rok'],tabela['Tydzień'])]
tabela=tabela.drop('Rok',1)
tabela=tabela.drop('Tydzień',1)
tabela.set_index("Date", inplace=True)
```

```python
t = tabela.groupby('Date').apply(lambda x: (x["Poziom Be7"]*x["Niepewność pomiarowa"].sum()).sum())
#tabela = tabela.resample('M',how = 'mean')
#tabela['srednia_wazona'] = tabela['poziom/niepewnosc']*tabela['Niepewność pomiarowa']/4
```
